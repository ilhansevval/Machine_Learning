```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("linear-regression-dataset.csv",sep = ";") #df değişkenine dataset verileri aktarılır.
```

```python
plt.scatter(df.experience,df.salary) #x ve y koordinatlarına hangi sütunların geleceği belirlenir.
plt.xlabel("experience") #x labelın adı atanır
plt.ylabel("salary")  #y labelın adı atanır.
plt.show()   #çizdirilir.
```

```python
#%% linear regression
# sklearn library
from sklearn.linear_model import LinearRegression

# linear regression model
linear_reg = LinearRegression()

x = df.experience.values.reshape(14,1) #experience sütunu fit edilmek için x değişkenine yeniden şekillendirilerek atandı.
y = df.salary.values.reshape(14,1)     #salary sütunu fit edilmek için y değişkenine yeniden şekillendirilerek atandı.

linear_reg.fit(x,y)                    #fit ediyoruz.
```

```python
#%% prediction
import numpy as np

b0 = linear_reg.predict([[0]]) #fit ettiğimiz linear regressiondaki 0. indis b0 değikenine atanır.
print("b0: ",b0)


```

```python
b0_ = linear_reg.intercept_  #β0:Doğrunun y-eksenini kestiği yer ve regresyon sabitidir intercept ile çağırılır.
print("b0_: ",b0_)   # intercept the y-axis

```

```python
b1 = linear_reg.coef_  #β1:Doğrunun eğimi veya regresyon katsayısıdır coef ile çağırılır.
print("b1: ",b1)   # slope

# salary = 1663 + 1138*experience 

```

```python
new_salary = 1663 + 1138*11 #b0=1663 b1=1138 11 yıllık deneyim hesaplanmış.
print(new_salary)

```

```python
b11 = linear_reg.predict([[11]])  #aynı işlemin linear regresyonda tahmin edilmiş hali
print("b11: ",b11)


```
